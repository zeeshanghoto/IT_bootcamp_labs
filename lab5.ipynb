```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
df
```

```python
y=df['Survived']
```

```python
X=df[['Sex','Age','Embarked','Pclass']]
```

```python
y.isnull().sum()
```

```python
X.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder

lb=LabelEncoder()
X['Embarked'] =lb.fit_transform(X['Embarked'])

lb1=LabelEncoder()
X['Sex'] =lb1.fit_transform(X['Sex'])
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X.values,y.values,random_state = 42,test_size= 0.2)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
clf1 = KNeighborsClassifier(n_neighbors = 7,p=1)#manhattan
clf2 = KNeighborsClassifier(n_neighbors = 3,p=2)#euclidean
```

```python
clf1.fit(X_train,y_train)

clf2.fit(X_train,y_train)
```

```python
pred1 = clf1.predict(X_test)
pred2 = clf2.predict(X_test)
```

```python
cm1 = confusion_matrix(y_test,pred1)
cm2 = confusion_matrix(y_test,pred2)
```

```python
print(cm1)
```

```python
print(cm2)
```

```python
cl1 = classification_report(y_test,pred1)
cl2 = classification_report(y_test,pred2)
```
