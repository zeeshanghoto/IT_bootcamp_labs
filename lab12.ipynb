```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, fpgrowth,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv("groceries - groceries.csv")
df
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame containing the transactional data
# Replace 'Item(s)' with the actual name of the column containing the items

# Step 1: Drop the 'Item(s)' column as it's not needed for the final format
df.drop('Item(s)', axis=1, inplace=True)

# Step 2: Create a new DataFrame to store the converted transactions
converted_data = {
    'TID': [],
    'Items': []
}

# Step 3: Iterate through each row and extract items
for index, row in df.iterrows():
    tid = f"T{index+1}"  # Generate the transaction ID T1, T2, ...
    items = [item for item in row if pd.notna(item)]  # Extract non-null items
    converted_data['TID'].append(tid)
    converted_data['Items'].append(items)

# Step 4: Create the final DataFrame in the desired format
converted_df = pd.DataFrame(converted_data)

# Step 5: Display the result
print(converted_df)
```

```python
te = TransactionEncoder()
te_ary = te.fit(converted_df['Items']).transform(converted_df['Items'])
df_trans = pd.DataFrame(te_ary, columns=te.columns_)

min_support = 0.01
min_confidence = 0.01

frequent_itemsets = fpgrowth(df_trans, min_support=min_support, use_colnames=True)

association_rules_df = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

print("Frequent Itemsets:")
display(frequent_itemsets)
print("\nAssociation Rules:")
display(association_rules_df)
```

```python
sorted_df = association_rules_df.sort_values(by=['support', 'confidence'], ascending=[False,False])

print(sorted_df)
```
