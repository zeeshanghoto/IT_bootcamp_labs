```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Plot the dataset
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Sample Dataset')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the dataset
model.fit(X, y)

# Get the coefficients (slope and intercept)
slope = model.coef_[0][0]
intercept = model.intercept_[0]

# Make predictions
y_pred = model.predict(X)

# Calculate the Mean Squared Error
mse = mean_squared_error(y, y_pred)

# Print the results
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"Mean Squared Error: {mse}")

# Plot the linear regression line
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
In
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('Salary_dataset.csv')

# Explore the dataset
print(data.head())  # Display the first few rows of the dataset

# Separate the feature 'YearsExperience' (X) and the target 'Salary' (y)
X = data['YearsExperience'].values.reshape(-1, 1)
y = data['Salary'].values

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the dataset
model.fit(X, y)

# Get the coefficients (slope and intercept)
slope = model.coef_[0]
intercept = model.intercept_

# Make predictions
y_pred = model.predict(X)

# Calculate the Mean Squared Error
mse = mean_squared_error(y, y_pred)

# Print the results
print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")
print(f"Mean Squared Error: {mse}")

# Plot the linear regression line
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Salary vs. Years of Experience')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('Salary_dataset.csv')

# Separate the feature 'YearsExperience' (X) and the target 'Salary' (y)
X = data['YearsExperience'].values.reshape(-1, 1)
y = data['Salary'].values

# Define the number of folds (K) for cross-validation
K = 5

# Create a Linear Regression model
model = LinearRegression()

# Initialize KFold with the number of splits
kf = KFold(n_splits=K, shuffle=True, random_state=42)

mse_scores = []  # To store Mean Squared Error for each fold

# Perform K-fold cross-validation
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Fit the model to the training data
    model.fit(X_train, y_train)

    # Make predictions on the test data
    y_pred = model.predict(X_test)

    # Calculate the Mean Squared Error for this fold
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

# Calculate the average Mean Squared Error across all folds
average_mse = np.mean(mse_scores)

# Print the results
print(f"Average Mean Squared Error: {average_mse}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, mean_squared_error

# Load the dataset
data = pd.read_csv('Salary_dataset.csv')

# Separate the feature 'YearsExperience' (X) and the target 'Salary' (y)
X = data['YearsExperience'].values.reshape(-1, 1)
y = data['Salary'].values

# Define the number of folds (K) for cross-validation
K = 5

# Create a Linear Regression model
model = LinearRegression()

# Initialize KFold with the number of splits
kf = KFold(n_splits=K, shuffle=True, random_state=42)

r2_scores = []  # To store R-squared for each fold
mse_scores = []  # To store Mean Squared Error for each fold

# Perform K-fold cross-validation
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Fit the model to the training data
    model.fit(X_train, y_train)

    # Make predictions on the test data
    y_pred = model.predict(X_test)

    # Calculate R-squared for this fold
    r2 = r2_score(y_test, y_pred)
    r2_scores.append(r2)

    # Calculate the Mean Squared Error for this fold
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

# Calculate the average R-squared and Mean Squared Error across all folds
average_r2 = np.mean(r2_scores)
average_mse = np.mean(mse_scores)

# Print the results
print(f"Average R-squared: {average_r2}")
print(f"Average Mean Squared Error: {average_mse}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('Salary_dataset.csv')

# Separate the feature 'YearsExperience' (X) and the target 'Salary' (y)
X = data['YearsExperience'].values.reshape(-1, 1)
y = data['Salary'].values

# Create a Linear Regression model
model = LinearRegression()

# Define the hyperparameters and their range of values to tune
param_grid = {
    'fit_intercept': [True, False]
}

# Create GridSearchCV with the model, parameter grid, and cross-validation strategy (K-fold)
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the GridSearchCV to the dataset to find the best hyperparameters
grid_search.fit(X, y)

# Get the best hyperparameters and the corresponding model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Print the best hyperparameters
print("Best Hyperparameters:")
print(best_params)

# Make predictions using the best model
y_pred = best_model.predict(X)

# Calculate the Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error with Best Model: {mse}")
```
